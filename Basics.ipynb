```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data for Figure 3
data_fig3 = {
    'TextRank': [36, 40.1, 38],
    'LSTM': [34.61, 38.15, 36.4],
    'Transformer': [38.14, 41.05, 39.63],
    'LSTM+TextRank': [36.54, 41.63, 40.11],
    'Transformer+TextRank': [39, 41.96, 41.92]
}

# Data for Figure 4
data_fig4 = {
    'LSTM+TextRank': [36.54, 41.63, 40.11],
    'Transformer+TextRank': [39, 41.96, 41.92],
    'LSTM+TextRank*': [36.66, 41.96, 40.52],
    'Transformer+TextRank*': [39.25, 42.32, 42.6]
}

# Create dataframes for each figure
df_fig3 = pd.DataFrame(data_fig3, index=['Rouge-1-f', 'Rouge-2-f', 'Rouge-L-f'])
df_fig4 = pd.DataFrame(data_fig4, index=['Rouge-1-f', 'Rouge-2-f', 'Rouge-L-f'])

# Function to wrap text in the column headers
def wrap_text_in_labels(labels, max_length=10):
    """Wrap text in labels to avoid overlap, breaking lines if necessary."""
    wrapped_labels = []
    for label in labels:
        words = label.split(' ')
        wrapped_label = '\n'.join(words) if len(label) > max_length else label
        wrapped_labels.append(wrapped_label)
    return wrapped_labels

# Plotting Figure 1
fig, ax = plt.subplots(figsize=(8, 6))  # Increase figure size for better clarity
df_fig3.plot(kind='bar', ax=ax)
ax.set_ylabel('ROUGE Score')
ax.set_title('Figure 1: ROUGE Values for the Five Methods')
ax.set_xticks([])  # Remove x-axis labels

# Adding grid lines
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

# Wrap the column headers text to avoid overlap
wrapped_columns = wrap_text_in_labels(df_fig3.columns, max_length=15)

# Making table clearer and preventing text overlap
table = ax.table(cellText=df_fig3.values, colLabels=wrapped_columns, rowLabels=df_fig3.index, loc='bottom', bbox=[0.15, -0.35, 0.7, 0.3])

# Customize table: Making it clearer
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(4.5, 4.5)  # Increase the size of the table

# Adjust row and column sizes to prevent overlap
for (i, j), cell in table.get_celld().items():
    cell.set_text_props(weight='normal', color='black')
    if i == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#40466e')
    elif j == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#40466e')
    else:
        cell.set_facecolor('#f1f1f2')
    cell.set_edgecolor('black')  # Cell border color
    cell.set_height(0.1)  # Increase row height for better visibility
    cell.set_width(0.15)  # Adjust column width for text

# Adjust layout to prevent overlap of x-axis and table
plt.subplots_adjust(bottom=0.4)  # Increase bottom margin for the table

plt.show()

# Plotting Figure 2
fig, ax = plt.subplots(figsize=(8, 6))  # Increase figure size for better clarity
df_fig4.plot(kind='bar', ax=ax)
ax.set_ylabel('ROUGE Score')
ax.set_title('Figure 2: ROUGE Values for the Four Methods')
ax.set_xticks([])  # Remove x-axis labels

# Adding grid lines
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.5)

# Wrap the column headers text to avoid overlap
wrapped_columns = wrap_text_in_labels(df_fig4.columns, max_length=15)

# Making table clearer and preventing text overlap
table = ax.table(cellText=df_fig4.values, colLabels=wrapped_columns, rowLabels=df_fig4.index, loc='bottom', bbox=[0.15, -0.35, 0.7, 0.3])

# Customize table: Making it clearer
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  # Increase the size of the table

# Adjust row and column sizes to prevent overlap
for (i, j), cell in table.get_celld().items():
    cell.set_text_props(weight='normal', color='black')
    if i == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#40466e')
    elif j == 0:
        cell.set_text_props(weight='bold', color='white')
        cell.set_facecolor('#40466e')
    else:
        cell.set_facecolor('#f1f1f2')
    cell.set_edgecolor('black')  # Cell border color
    cell.set_height(0.1)  # Increase row height for better visibility
    cell.set_width(0.15)  # Adjust column width for text

# Adjust layout to prevent overlap of x-axis and table
plt.subplots_adjust(bottom=0.4)  # Increase bottom margin for the table

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data: Models and their corresponding ROUGE scores (from the provided results)
models = ['Google Pegasus', 'BART', 'led-base-book-summary', 'bart-conversation-summary', 'Meeting Summary']
rouge_1_scores = [0.35, 0.26, 0.16, 0.16, 0.14]
rouge_2_scores = [0.25, 0.15, 0.05, 0.045, 0.04]
rouge_l_scores = [0.35, 0.26, 0.15, 0.16, 0.14]

# Convert the data into a pandas DataFrame for easier handling
data = {
    'Model': models,
    'ROUGE-1': rouge_1_scores,
    'ROUGE-2': rouge_2_scores,
    'ROUGE-L': rouge_l_scores
}

df = pd.DataFrame(data)

# Plotting the ROUGE scores for all models
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar width and positions for bars
bar_width = 0.2
index = np.arange(len(models))

# Plot the bars for each ROUGE metric
bar1 = ax.bar(index, df['ROUGE-1'], bar_width, label='ROUGE-1', color='b')
bar2 = ax.bar(index + bar_width, df['ROUGE-2'], bar_width, label='ROUGE-2', color='g')
bar3 = ax.bar(index + 2*bar_width, df['ROUGE-L'], bar_width, label='ROUGE-L', color='r')

# Add labels, title, and legend
ax.set_xlabel('Summarization Models')
ax.set_ylabel('ROUGE Scores')
ax.set_title('Comparison of ROUGE Scores for Different Summarization Models')
ax.set_xticks(index + bar_width)
ax.set_xticklabels(models)
ax.legend()

# Show the plot
plt.tight_layout()
plt.savefig("comparison_rouge_scores.png")  # Save the figure as a .png image
plt.show()
```
